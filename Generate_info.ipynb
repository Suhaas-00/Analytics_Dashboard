```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Configuration
NUM_USERS = 2000
NUM_DAYS = 60
MAX_SESSIONS_PER_USER = 5

sources = ["Organic", "Social", "Direct", "Referral", "Email"]
mediums = {
    "Organic": "Google",
    "Social": "Instagram",
    "Direct": "Website",
    "Referral": "PartnerSite",
    "Email": "Newsletter"
}

countries = ["India", "USA", "UK", "Canada", "Australia"]
devices = ["Mobile", "Desktop", "Tablet"]
pages = ["Home", "Product", "Blog", "Pricing", "Checkout", "Contact"]

start_date = datetime(2026, 1, 1)

data = []
session_counter = 1

for user_id in range(1, NUM_USERS + 1):
    
    num_sessions = random.randint(1, MAX_SESSIONS_PER_USER)
    session_dates = sorted([
        start_date + timedelta(days=random.randint(0, NUM_DAYS - 1))
        for _ in range(num_sessions)
    ])
    
    for session_index, session_date in enumerate(session_dates):
        
        session_id = f"S{session_counter}"
        session_counter += 1
        
        source = random.choice(sources)
        medium = mediums[source]
        country = random.choice(countries)
        device = random.choices(devices, weights=[0.7, 0.25, 0.05])[0]
        
        journey_length = random.randint(1, 4)
        journey_pages = random.sample(pages, journey_length)
        
        converted = 1 if "Checkout" in journey_pages and random.random() < 0.5 else 0
        
        user_type = "New" if session_index == 0 else "Returning"
        
        for step, page in enumerate(journey_pages, start=1):
            
            engagement_time = random.randint(20, 300)
            
            data.append([
                session_date,
                f"U{user_id}",
                session_id,
                source,
                medium,
                country,
                device,
                page,
                step,
                engagement_time,
                converted,
                user_type
            ])

df = pd.DataFrame(data, columns=[
    "Date",
    "User_ID",
    "Session_ID",
    "Source",
    "Medium",
    "Country",
    "Device",
    "Page",
    "Page_Order",
    "Engagement_Time",
    "Converted",
    "User_Type"
])

df = df.sort_values(by=["Date", "Session_ID", "Page_Order"])

df.to_csv("analytics_dataset.csv", index=False)

print("Professional dataset created successfully.")
print(df.head())
```
